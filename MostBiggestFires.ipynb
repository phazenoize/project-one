```python
# Dependencies and setup
import csv
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
from pathlib import Path
  
plt.close("all")
```

```python
# Load the CSV file 
wildfires_path = Path("Us_Lightning_Forest_Fires.csv")
wildfires_df = pd.read_csv(wildfires_path)
wildfires_df.head()
```

```python
# Group the data by "STATE" and calculate the total number of fires for each state
fires_by_state = wildfires_df['STATE'].value_counts()
```

```python
# Group the data by "STATE" and calculate the maximum fire size for each state
max_fire_size_by_state = wildfires_df.groupby('STATE')['FIRE_SIZE'].max()
```

```python
# Create a bar plot to visualize the maximum fire size by state maximum fire size by state
plt.figure(figsize=(12, 6))
max_fire_size_by_state.plot(kind='bar', color='red')
plt.title('Maximum Fire Size by State', fontsize=14)
plt.xlabel('State', fontsize=10)
plt.ylabel('Maximum Fire Size (acres)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Create a bar plot to visualize the maximum fire size by state
plt.figure(figsize=(12, 6))
fires_by_state.plot(kind='bar', color='grey')
plt.title('Total Number of Fires by State', fontsize=14)
plt.xlabel('State', fontsize=10)
plt.ylabel('Number of Fires', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
