```python
#dependencies and Setup
import csv
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
from pathlib import Path

plt.close("all")


# Load the CSV file 
wildfires_path = Path("Us_Lightning_Forest_Fires.csv")
wildfires_df = pd.read_csv(wildfires_path)
wildfires_df.head()
```

```python
# Filter the data for Arizona
arizona_df = wildfires_df[wildfires_df["STATE"] == "AZ"]
# Calculate the average fire extinguishing time by year
average_extinguishing_time_az = arizona_df.groupby("FIRE_YEAR")["Days_to_extinguish_fire"].mean()

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(average_extinguishing_time_az.index, average_extinguishing_time_az.values, marker='o', color='green')
plt.xlabel('Year')
plt.ylabel('Average Fire Extinguishing Time (Days)')
plt.title('Average Fire Extinguishing Time Over the Years in Arizona')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
