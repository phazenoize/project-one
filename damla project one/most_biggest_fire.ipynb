```python

#dependencies and setup
import csv
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
from pathlib import Path
  
plt.close("all")

```

```python
# Load the CSV file 
wildfires_path = Path("Us_Lightning_Forest_Fires.csv")
wildfires_df = pd.read_csv(wildfires_path)
wildfires_df.head()
```

```python
#Group the data by "State" and calculate the total nummber of fires for each state
fires_by_state = wildfires_df["STATE"].value_counts()
#state sutununu gruplandirir ve her eyaletin toplam yangin sayisini hesapliyor bu islem sonunda her bir eyaletin yangin sayisi o eyaletin adi ile birlikte bir pandas serisi olarak elde edilir
```

```python
# Group the data by "STATE" and calculate the maximum fire size for each state
max_fire_size_by_state = wildfires_df.groupby('STATE')['FIRE_SIZE'].max()
#her eyaletinin maksimum yangin boyutunu hesaplar
```

```python
#Bu kod, her eyaletteki toplam yangın sayısını ve her eyaletteki en büyük yangınların boyutunu gösteren iki ayrı bar plot oluşturacaktır. Mavi çubuklar, eyaletlerdeki toplam yangın sayısını gösterirken, turuncu çubuklar eyaletlerdeki en büyük yangınların boyutunu gösterecektir.

#Bu grafiklerle, hangi eyaletlerde en çok ve en büyük yangınların olduğunu kolayca görebilirsiniz.
```

```python
# Create a bar plot to visualize the total number of fires by state
plt.figure(figsize=(12, 6))
fires_by_state.plot(kind='bar', color='skyblue')
plt.title('Total Number of Fires by State', fontsize=16)
plt.xlabel('State', fontsize=14)
plt.ylabel('Number of Fires', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Create a bar plot to visualize the maximum fire size by state
plt.figure(figsize=(12, 6))
max_fire_size_by_state.plot(kind='bar', color='orange')
plt.title('Maximum Fire Size by State', fontsize=16)
plt.xlabel('State', fontsize=14)
plt.ylabel('Maximum Fire Size (acres)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Filter data for Arizona
arizona_df = wildfires_df[wildfires_df["STATE"] == "AZ"]

# Sort the data by FIRE_SIZE in descending order and select the top 5 rows
biggest_fires = arizona_df.nlargest(5, "FIRE_SIZE")

# Plot bar chart for the top 5 biggest fires in Arizona
plt.figure(figsize=(10, 6))
plt.bar(biggest_fires["Fire_Date"], biggest_fires["FIRE_SIZE"], color='salmon')
plt.xlabel('Fire Date')
plt.ylabel('Fire Size')
plt.title('Top 5 Biggest Fires in Arizona')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Filter data for Minnesota
minnesota_df = wildfires_df[wildfires_df["STATE"] == "MN"]

# Sort the data by FIRE_SIZE in descending order and select the top 5 rows
biggest_fires_mn = minnesota_df.nlargest(5, "FIRE_SIZE")

# Create a scatter plot for the top 5 biggest fires in Minnesota
plt.figure(figsize=(10, 6))
plt.scatter(biggest_fires_mn["Fire_Date"], biggest_fires_mn["FIRE_SIZE"], c=biggest_fires_mn["FIRE_SIZE"], cmap='YlOrRd', s=1000)
plt.colorbar(label="Fire Size")
plt.xlabel('Fire Date')
plt.ylabel('Fire Size')
plt.title('Top 5 Biggest Fires in Minnesota')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```
