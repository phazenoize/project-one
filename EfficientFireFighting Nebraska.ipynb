```python
# Dependencies and Setup
import csv
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
from pathlib import Path

plt.close("all")


# Load the CSV file 
wildfires_path = Path("Us_Lightning_Forest_Fires.csv")
wildfires_df = pd.read_csv(wildfires_path)
wildfires_df.head()
```

```python
# Filter the data for Nebraska (NE) and select the necessary columns
nebraska_df = wildfires_df[wildfires_df["STATE"] == "NE"][["FIRE_YEAR", "Fire_Date", "Days_to_extinguish_fire"]]

# Convert the "Fire_Date" column to datetime data type
nebraska_df["Fire_Date"] = pd.to_datetime(nebraska_df["Fire_Date"])

# Calculate the total average fire extingished by month and fill missing values with 0
monthly_days_to_extinguish_fire_nebraska = nebraska_df.groupby(nebraska_df["Fire_Date"].dt.month)["Days_to_extinguish_fire"].mean().reindex(range(1, 13), fill_value=0)

# Define the month names
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(months, monthly_days_to_extinguish_fire_nebraska, color='grey')
plt.xlabel('Month')
plt.ylabel('Average Extinguish Time')
plt.title('Nebraska Wildfires: Average Extinguish by Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
